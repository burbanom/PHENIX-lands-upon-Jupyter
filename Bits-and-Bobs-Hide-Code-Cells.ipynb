Oftentimes you will find that your report looks nicer if you can hide the source code. Adding the cell below at the begining of your notebook will provide you with a toggle switch to hide/show the source code cells. This solution comes from: http://stackoverflow.com/questions/27934885/how-to-hide-code-from-cells-in-ipython-notebook-visualized-with-nbviewer

```python
from IPython.display import HTML

HTML('''<script>
code_show=true; 
function code_toggle() {
 if (code_show){
 $('div.input').hide();
 } else {
 $('div.input').show();
 }
 code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
<form action="javascript:code_toggle()"><input type="submit" value="Click here to toggle on/off the raw code."></form>''')
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
def a_funct(x):
    return x**2
```

```python
x = np.arange(10)
```

```python
y = a_funct(x)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

The code cells used to generate the plot shown above can now be easily hidden.
